```python
import numpy as np
import pandas as pd
#importing the train_test_split
from sklearn.model_selection import train_test_split
#importing the accuracy score
from sklearn.metrics import accuracy_score
#importing the seaborne and matpotlib
import seaborn as sns
import matplotlib.pyplot as plt
#importing the standard
from sklearn.preprocessing import StandardScaler
```

```python
class logistic_regression():
  #to initialise the hyperparameters
  def __init__(self,no_of_iteration,learning_rate):

    self.learning_rate = learning_rate

    self.no_of_iteration = no_of_iteration

  #to train the data
  def fit(self,X,Y):
    self.m,self.n = X.shape
    self.w = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y

    for i in range(self.no_of_iteration):
      self.update_weights(X)

  def update_weights(self,X):

    y_hat = 1/(1+np.exp(-(self.X.dot(self.w)+self.b)))

    dw = 1*(np.dot((y_hat-self.Y).T,self.X))/self.m
    db = 1*(np.sum(y_hat-self.Y))/self.m

    self.w=self.w-self.learning_rate*dw
    self.b=self.b-self.learning_rate*db


  def predict(self,X):
    y_prediction = 1/(1+np.exp(-(X.dot(self.w)+self.b)))

    y_pred=np.where(y_prediction>0.5,1,0)

    return y_pred
```

```python
data = pd.read_csv("/content/diabetes.csv")
#dropping the Unnamed column


X = data.drop(columns="Outcome",axis=1)

Y = data["Outcome"]

scaler= StandardScaler()

scaler.fit(X)
standardized_data = scaler.transform(X)
X = standardized_data

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)

model = logistic_regression(no_of_iteration=1000,learning_rate=0.01)
model.fit(X_train,Y_train)

y_predictions = model.predict(X_train)

#predicting the accuracy on the training data
training_data_accuracy = accuracy_score(Y_train,y_predictions)
print("Accuracy on training data is : ",training_data_accuracy*100)

# predicting the data on the test data
y_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,y_prediction)
print("Accuracy on the testing data is: ", test_data_accuracy*100)
```
